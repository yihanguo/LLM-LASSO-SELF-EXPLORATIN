```python
def is_numeric_like(col: pd.Series, thresh: float = 0.95) -> bool:
    """
    Return True if, after stripping commas/spaces, at least `thresh`
    fraction of non-null values can be parsed as numbers.
    """
    if col.dtype != object:
        return False
    
    cleaned = (
        col.astype(str)
           .str.replace(r"[,\s]", "", regex=True)
           .replace({"": np.nan})
    )
    numeric = pd.to_numeric(cleaned, errors="coerce")
    return numeric.notna().mean() >= thresh
obj_cols = spotify.select_dtypes(exclude="number").columns.tolist()
numeric_like = [c for c in obj_cols if is_numeric_like(spotify[c])]
textual_cols = [c for c in obj_cols if c not in numeric_like]
print("Detected numeric-like (string → numeric):")
print(numeric_like, "\n")
print("Remaining textual columns:")
print(textual_cols, "\n")
```

```python
def auto_numeric(df, thresh=0.90):
    """
    Convert object columns that are ≥ `thresh` fraction numeric-like
    (after stripping commas/space) to float.
    """
    obj_cols = df.select_dtypes(exclude="number").columns
    for c in obj_cols:
        cleaned = df[c].astype(str).str.replace(r"[,\s]", "", regex=True)
        ok_ratio = cleaned.str.fullmatch(r"\d+(\.\d+)?").mean()
        if ok_ratio >= thresh:
            df[c] = pd.to_numeric(cleaned, errors="coerce")
    return df

spotify_auto = auto_numeric(spotify.copy(), thresh=0.90)
print(spotify_auto.dtypes)
```

```python
import pandas as pd
import numpy as np
import re

########################################################################
# 0 – (Re)load and normalise column names
########################################################################
spotify = pd.read_csv("Most Streamed Spotify Songs 2024.csv",
                      encoding="ISO-8859-1")

spotify.columns = (
    spotify.columns
          .str.strip()          # remove leading/trailing spaces
          .str.replace(" +", " ", regex=True)  # squash double spaces
)

########################################################################
# 1 – columns to keep as text
########################################################################
keep_text = ["Track", "Album Name", "Artist", "ISRC"]

########################################################################
# 2 – Release Date  → numeric offset
########################################################################
if "Release Date" in spotify.columns:
    spotify["Release Date"] = pd.to_datetime(
        spotify["Release Date"], errors="coerce", infer_datetime_format=True
    )

    origin = spotify["Release Date"].min()
    spotify["release_days"] = (spotify["Release Date"] - origin).dt.days

    # drop only once; harmless if repeated because we guard with `if`
    spotify.drop(columns=["Release Date"], inplace=True, errors="ignore")

########################################################################
# 3 – helper to coerce numeric-like object columns
########################################################################
def force_numeric(series):
    return (
        series.astype(str)
              .str.replace(r"[,\s]", "", regex=True)
              .replace({"": np.nan, "N/A": np.nan, "—": np.nan})
              .astype(float)
    )

########################################################################
# 4 – convert everything except the text columns
########################################################################
for col in spotify.columns:
    if col in keep_text:
        continue
    if spotify[col].dtype == "object":
        spotify[col] = force_numeric(spotify[col])

########################################################################
# 5 – verify
########################################################################
print(spotify.dtypes)
```

```python
import pandas as pd
import numpy as np

###############################################################################
# 1️⃣  Identify numeric columns
###############################################################################
num_cols = spotify.select_dtypes(include=["number"]).columns

###############################################################################
# 2️⃣  Copy the numeric part so the original DataFrame stays unchanged
###############################################################################
num_filled = spotify[num_cols].copy()

###############################################################################
# 3️⃣  Column-wise imputation
###############################################################################
for col in num_cols:
    col_data = num_filled[col]
    if col_data.notna().any():                 # at least one non-NaN
        median_val = col_data.median()
        num_filled[col] = col_data.fillna(median_val)
    else:                                      # column is fully NaN
        num_filled[col] = 0.0                  # or int(0) if you prefer

###############################################################################
# 4️⃣  Summary stats  (mean / median / std) after imputation
###############################################################################
summary = num_filled.agg(['mean', 'median', 'std']).T.round(2)
display(summary)

###############################################################################
# 5️⃣  (Optional) merge back into a full DataFrame
###############################################################################
spotify_imputed = spotify.copy()
spotify_imputed[num_cols] = num_filled
```

```python
import pandas as pd, numpy as np
from pathlib import Path

###############################################################################
# helper functions (unchanged, except coerce_numeric now returns df always)
###############################################################################
def load_csv(path, sep=","):
    try:
        return pd.read_csv(path, sep=sep)
    except UnicodeDecodeError:
        return pd.read_csv(path, sep=sep, encoding="ISO-8859-1")

def coerce_numeric(df):
    for col in df.columns:
        coerced = pd.to_numeric(
            df[col].astype(str).str.replace(r"[,\s]", "", regex=True),
            errors="coerce"
        )
        if coerced.notna().any():
            df[col] = coerced.astype(float)
    return df

def impute_median_or_zero(df):
    for col in df.select_dtypes(include="number"):
        if df[col].notna().any():
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = 0.0
    return df

def summarize(df):
    num = df.select_dtypes(include="number")
    if num.empty:
        return None
    return num.agg(['mean', 'median', 'std']).T.round(2)

###############################################################################
# CSV paths **plus the proper delimiter** for each file
###############################################################################
csv_paths = {
    # path                           # delimiter
    "bank-full"            : ("bank+marketing/bank/bank-full.csv",           ";"),
    "bank"                 : ("bank+marketing/bank/bank.csv",                ";"),
    "bank-additional-full" : ("bank+marketing/bank-additional/bank-additional-full.csv", ";"),
    "bank-additional"      : ("bank+marketing/bank-additional/bank-additional.csv",       ";"),

    "TCGA_Mutations_all"   : ("glioma+grading+clinical+and+mutation+features+dataset/TCGA_GBM_LGG_Mutations_all.csv", ","),
    "TCGA_InfoWithGrade"   : ("glioma+grading+clinical+and+mutation+features+dataset/TCGA_InfoWithGrade.csv",         ","),

    "winequality-red"      : ("wine+quality/winequality-red.csv",   ";"),
    "winequality-white"    : ("wine+quality/winequality-white.csv", ";"),

    "diabetes"             : ("diabetes.csv", ","),

    "spotify"              : ("Most Streamed Spotify Songs 2024.csv", ","),
}

###############################################################################
# process each file
###############################################################################
summaries, out_dir = {}, Path("numeric_summaries")
out_dir.mkdir(exist_ok=True)

for name, (path_str, sep) in csv_paths.items():
    path = Path(path_str)
    print(f"\n🔹 Processing {name:<25}  sep='{sep}'")

    if not path.exists():
        print("   ⚠️  File not found, skipping.")
        continue

    df = load_csv(path, sep=sep)
    df = impute_median_or_zero(coerce_numeric(df))
    summary = summarize(df)

    if summary is None:
        print("   ℹ️  No numeric columns detected.")
        continue

    display(summary.head())
    summary.to_csv(out_dir / f"{name}_summary.csv")
    summaries[name] = summary
    print(f"   ✅ saved → {out_dir / (name + '_summary.csv')}")
```

```python
import pandas as pd
import numpy as np
import warnings
from pathlib import Path

###############################################################################
# ▸ 1.  Silence the specific NumPy warning globally
###############################################################################
np.seterr(invalid="ignore")  # suppress RuntimeWarning: invalid value encountered in subtract

# If you prefer to keep other warnings intact, you can also wrap the summary step
# in a local context manager (see comment further below).

###############################################################################
# ▸ 2.  Helper functions
###############################################################################
def load_csv(path, sep=","):
    """Read CSV with fallback encoding."""
    try:
        return pd.read_csv(path, sep=sep)
    except UnicodeDecodeError:
        return pd.read_csv(path, sep=sep, encoding="ISO-8859-1")

def coerce_numeric(df):
    """Turn numeric-like object columns into float; leave true text unchanged."""
    for col in df.columns:
        coerced = pd.to_numeric(
            df[col].astype(str).str.replace(r"[,\s]", "", regex=True),  # strip commas/space
            errors="coerce",
        )
        if coerced.notna().any():        # at least one real number
            df[col] = coerced.astype(float)
    return df

def impute_median_or_zero(df):
    """
    • If a numeric column has ≥1 non-NaN → fill NaNs with column median.
    • If it is entirely NaNs → fill with 0.
    """
    for col in df.select_dtypes(include="number"):
        if df[col].notna().any():
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = 0.0
    return df

def summarize(df):
    """Return mean/median/std for every numeric column (rounded)."""
    num = df.select_dtypes(include="number").fillna(0)  # final sweep: no NaNs left
    if num.empty:
        return None
    # Optionally silence warnings only here:
    # with warnings.catch_warnings():
    #     warnings.filterwarnings("ignore", category=RuntimeWarning, message="invalid value encountered")
    #     return num.agg(['mean', 'median', 'std']).T.round(2)
    return num.agg(['mean', 'median', 'std']).T.round(2)

###############################################################################
# ▸ 3.  CSV paths + proper delimiter (same list as before)
###############################################################################
csv_paths = {
    "bank-full"            : ("bank+marketing/bank/bank-full.csv",           ";"),
    "bank"                 : ("bank+marketing/bank/bank.csv",                ";"),
    "bank-additional-full" : ("bank+marketing/bank-additional/bank-additional-full.csv", ";"),
    "bank-additional"      : ("bank+marketing/bank-additional/bank-additional.csv",       ";"),
    "TCGA_Mutations_all"   : ("glioma+grading+clinical+and+mutation+features+dataset/TCGA_GBM_LGG_Mutations_all.csv", ","),
    "TCGA_InfoWithGrade"   : ("glioma+grading+clinical+and+mutation+features+dataset/TCGA_InfoWithGrade.csv",         ","),
    "winequality-red"      : ("wine+quality/winequality-red.csv",   ";"),
    "winequality-white"    : ("wine+quality/winequality-white.csv", ";"),
    "diabetes"             : ("diabetes.csv", ","),
    "spotify"              : ("Most Streamed Spotify Songs 2024.csv", ","),
}

###############################################################################
# ▸ 4.  Process each dataset and store summaries
###############################################################################
summaries, out_dir = {}, Path("numeric_summaries")
out_dir.mkdir(exist_ok=True)

for name, (path_str, sep) in csv_paths.items():
    path = Path(path_str)
    print(f"\n🔹 Processing {name:<25}  sep='{sep}'")
    if not path.exists():
        print("   ⚠️  File not found, skipping.")
        continue

    df = load_csv(path, sep=sep)
    df = impute_median_or_zero(coerce_numeric(df))

    summary = summarize(df)
    if summary is None:
        print("   ℹ️  No numeric columns detected.")
        continue

    display(summary.head())  # show a preview inside Jupyter
    summary.to_csv(out_dir / f"{name}_summary.csv")
    summaries[name] = summary
    print(f"   ✅ saved → {out_dir / (name + '_summary.csv')}")

print("\n🎉  All done!")
```
